```python
new_list = list(input("Enter a string: "))

if len(new_list) >= 2:
    del new_list[:2]


rev_str = new_list[::-1]

print("".join(rev_str))
```

```python
def replace_python(sen):
    return sen.replace('python', 'pythons')

sen = input("Enter a sentence: ")

mod_sen = replace_python(sen)

print("Modified sentence:", mod_sen)
```

```python
def cal_grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 80:
        return 'B'
    elif marks >= 70:
        return 'C'
    elif marks >= 60:
        return 'D'
    else:
        return 'F'

marks = float(input("Enter the class score: "))

grade = cal_grade(marks)

print("The letter grade is:", grade)
```
